```python
def monotoneIncreasing(array):
    l = len(array)
    for i in range(l-1):
        if (array[i] < array[i+1]):
            continue
        else:
            return False
    return True

def monotoneDecreasing(array):
    l = len(array)
    for i in range(l-1):
        if (array[i] > array[i+1]):
            continue
        else:
            return False
    return True


if __name__ == "__main__":
    array1 = [1,2,3,4,5]
    array2 = [5,4,3,2,1,0]
    print(array1)
    print("is increasing ? ", monotoneIncreasing(array1))
    print(array2)
    print("is decreasing ? ", monotoneDecreasing(array2))
    
    
```
